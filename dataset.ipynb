```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./extention_of_Z-Alizadeh_sani_dataset.csv')
df.head()
```

```python
float_columns = df.select_dtypes('float64').columns
for column in float_columns:
    df[column] = pd.cut(x = df[column], bins= 3, labels= range(3), right= False)

df.head()
```

```python
int_columns = df.select_dtypes('int64').columns
df[int_columns] = df.select_dtypes('int64').astype('category')
```

```python
df.nunique()
```

```python
new_dataset = pd.get_dummies(data=df)
new_dataset.head()
```

```python
new_dataset = new_dataset.drop(['Cath_CAD', 'Cath_Normal'], axis= 1)
new_dataset.columns
```

```python
new_dataset = pd.concat((new_dataset, df.Cath), axis= 1)
new_dataset.head()
```

```python
new_dataset.to_csv('./binarized.csv', index=False)
```
